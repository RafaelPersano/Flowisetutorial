```python
import random

def gerar_jogos():
    mercados = ["Over 1.5", "Over 0.5 HT", "Casa ganha", "Mais ataques time favorito", "Gol no 1º tempo"]
    return [
        {"id": i, "mercado": random.choice(mercados), "odd": round(random.uniform(1.5, 1.7), 2)}
        for i in range(1, 11)
    ]

def escolher_combinadas(jogos, num_combinadas):
    return [random.sample(jogos, 2) for _ in range(num_combinadas)]

def simular_resultado(odd):
    return random.random() < (odd - 1) / 2

def estrategia_recuperacao(perdas, banca):
    return min(5 + perdas * 1.5, banca * 0.1)

def simular_dia(banca, num_combinadas, perdas):
    jogos = gerar_jogos()
    combinadas = escolher_combinadas(jogos, num_combinadas)
    ganho_dia = 0

    for i, aposta in enumerate(combinadas, 1):
        valor_aposta = estrategia_recuperacao(perdas, banca)
        banca -= valor_aposta

        resultado = all(simular_resultado(jogo["odd"]) for jogo in aposta)
        if resultado:
            ganho = valor_aposta * aposta[0]["odd"] * aposta[1]["odd"]
            banca += ganho
            ganho_dia += ganho - valor_aposta
            perdas = 0
            print(f"\nAposta {i}: ✅ Ganhou {ganho:.2f} BRL")
        else:
            perdas += valor_aposta
            print(f"\nAposta {i}: ❌ Perdeu {valor_aposta:.2f} BRL")

        print(f"💰 Banca atual: {banca:.2f} BRL")

    return banca, perdas

def simular_30_dias(banca_inicial, num_combinadas):
    banca = banca_inicial
    perdas = 0

    for dia in range(1, 31):
        print(f"\n📅 Dia {dia} 🏆")
        banca, perdas = simular_dia(banca, num_combinadas, perdas)

        if banca <= 0:
            print("🔥 Banca zerada! Jogo encerrado.")
            break

    print(f"\n🏁 Fim dos 30 dias! Banca final: {banca:.2f} BRL")

# Rodando o simulador
banca_inicial = float(input("Digite sua banca inicial: "))
num_combinadas = int(input("Escolha o número de combinadas (1, 2 ou 3): "))
simular_30_dias(banca_inicial, num_combinadas)

```

```python
import random

def simulate_betting(initial_bankroll, days, bets_per_day, base_bet_percentage, win_chance_range, odds_range):
    bankroll = initial_bankroll
    history = []

    for day in range(1, days + 1):
        day_result = {"day": day, "bets": [], "bankroll": bankroll}

        for _ in range(bets_per_day):
            bet_amount = round(bankroll * base_bet_percentage, 2)
            if bet_amount < 0.01:
                break  # Stop if bankroll is too low

            win_chance = random.uniform(*win_chance_range)
            odds = random.uniform(*odds_range)

            if random.random() < win_chance:
                winnings = round(bet_amount * (odds - 1), 2)
                bankroll += winnings
                day_result["bets"].append({"result": "✅ Ganhou", "amount": winnings})
            else:
                bankroll -= bet_amount
                day_result["bets"].append({"result": "❌ Perdeu", "amount": -bet_amount})

        bankroll = round(bankroll, 2)
        day_result["bankroll"] = bankroll
        history.append(day_result)

    return history

# Parâmetros ajustados para um desempenho mais realista
initial_bankroll = 100
num_days = 30
bets_per_day = 2
base_bet_percentage = 0.05  # 5% da banca por aposta
win_chance_range = (0.4, 0.6)  # 40% a 60% de chance de acerto
odds_range = (1.8, 2.5)  # Odds entre 1.8 e 2.5

betting_history = simulate_betting(initial_bankroll, num_days, bets_per_day, base_bet_percentage, win_chance_range, odds_range)

# Exibir resultado final
for day_data in betting_history:
    print(f"\n📅 Dia {day_data['day']} 🏆")
    for bet in day_data["bets"]:
        print(f"Aposta: {bet['result']} {bet['amount']} BRL")
    print(f"💰 Banca atual: {day_data['bankroll']} BRL")

print(f"\n🏁 Fim dos {num_days} dias! Banca final: {betting_history[-1]['bankroll']} BRL")
```

```python
import random

def gerar_odds():
    """Gera odds aleatórias entre 1.8 e 2.5."""
    return round(random.uniform(1.8, 2.5), 2)

def definir_resultado():
    """Define se uma aposta é vencedora ou perdedora com 40% a 60% de chance de acerto."""
    return random.uniform(0, 1) < random.uniform(0.4, 0.6)

def calcular_retorno(odds, aposta):
    """Calcula o retorno da aposta vencedora."""
    retorno = aposta * odds
    return retorno

def simular_apostas():
    banca = float(input("Digite a banca inicial: "))
    while banca <= 0:
        banca = float(input("Banca inválida. Digite um valor positivo: "))

    while banca > 0:
        num_eventos = random.choice([2, 3])  # Dupla ou tripla
        odds = [gerar_odds() for _ in range(num_eventos)]
        odd_total = round(sum(odds) / num_eventos, 2)
        aposta = round(banca * 0.05, 2)

        if definir_resultado():  # Aposta vencedora
            ganho = calcular_retorno(odd_total, aposta)
            banca += ganho - aposta
            print(f"✅ Aposta vencedora! Odds: {odds}, Ganhou R$ {ganho:.2f}, Nova banca: R$ {banca:.2f}")
        else:
            banca -= aposta
            print(f"❌ Aposta perdida! Odds: {odds}, Perdeu R$ {aposta:.2f}, Nova banca: R$ {banca:.2f}")

        if banca <= 0:
            print("💸 Banca zerada! Fim da simulação.")
            break

if __name__ == "__main__":
    simular_apostas()
```

```python
import random

# Configuração inicial
banca_inicial = 1000  # Define a banca inicial
banca = banca_inicial

dias_simulados = 30
lucro_diario_max = 0.05  # 5% de lucro
perda_diaria_max = 0.10  # 10% de perda

# Função para simular uma aposta múltipla
def aposta_multipla():
    tipo = random.choice(["dupla", "tripla"])
    odds = [round(random.uniform(1.8, 2.5), 2) for _ in range(2 if tipo == "dupla" else 3)]
    odd_total = round(sum(odds) / len(odds), 2)
    resultado = random.random() < (1 / odd_total)  # Probabilidade inversa da odd total
    return odd_total, resultado

# Simulação de 30 dias
for dia in range(1, dias_simulados + 1):
    banca_inicio_dia = banca
    print(f"\nDia {dia} - Banca Inicial: R${banca:.2f}")

    while True:
        odd, venceu = aposta_multipla()
        aposta_valor = banca * 0.05  # 5% da banca por aposta

        if venceu:
            ganho = aposta_valor * (odd - 1)
            banca += ganho
        else:
            banca -= aposta_valor

        lucro_diario = (banca - banca_inicio_dia) / banca_inicio_dia

        print(f"  Aposta: R${aposta_valor:.2f} | Odd: {odd} | {'✅ Ganhou' if venceu else '❌ Perdeu'} | Banca: R${banca:.2f}")

        if lucro_diario >= lucro_diario_max:
            print(f"  🚀 Meta diária de lucro atingida! ({lucro_diario * 100:.2f}%)")
            break
        if lucro_diario <= -perda_diaria_max:
            print(f"  ⛔ Stop Loss atingido! ({lucro_diario * 100:.2f}%)")
            break

    if banca <= 0:
        print("💀 Banca zerada! Fim da simulação.")
        break

print(f"\n🏁 Simulação encerrada! Banca final: R${banca:.2f}")
```

```python
import random

def simular_apostas(banca_inicial, dias):
    """
    Simula apostas com uma taxa de vitória de 60% por um determinado número de dias.

    Args:
        banca_inicial (float): A banca inicial.
        dias (int): O número de dias para simular.

    Returns:
        list[float]: Uma lista contendo o valor da banca ao final de cada dia.
    """
    banca = banca_inicial
    historico = []

    for dia in range(1, dias + 1):
        banca_inicio_dia = banca
        lucro_dia = 0
        aposta_inicial = banca * 0.05  # 5% da banca
        print(f"\nDia {dia} - Banca Inicial: R${banca:.2f}")

        while True:
            aposta = min(aposta_inicial, banca)
            if aposta < 0.01:  # Stop if bet amount is too low
                print("  Aposta mínima não atingida. Encerrando o dia.")
                break

            odd = round(random.uniform(1.8, 2.5), 2)

            # Ajustando a probabilidade de vitória para 60%
            resultado = "✅ Ganhou" if random.random() < 0.6 else "❌ Perdeu"

            if resultado == "✅ Ganhou":
                ganho = aposta * (odd - 1)
                banca += ganho
                lucro_dia += ganho
            else:
                banca -= aposta
                lucro_dia -= aposta

            print(f"  Aposta: R${aposta:.2f} | Odd: {odd} | {resultado} | Banca: R${banca:.2f}")

            if lucro_dia >= banca_inicio_dia * 0.05:
                print("  ✅ Stop Gain atingido! (+5%)")
                break
            if lucro_dia <= -banca_inicio_dia * 0.06:
                print("  ⛔ Stop Loss atingido! (-6%)")
                break

        historico.append(banca)
        if banca <= 0:
            print("  ❌ Banca zerada! Fim da simulação.")
            break

    return historico

# Entrada do usuário
banca_inicial = float(input("Digite a banca inicial: "))
dias = int(input("Digite o número de dias da simulação: "))

# Executar simulação
historico_banca = simular_apostas(banca_inicial, dias)

# Exibir o histórico da banca
print("\nHistórico da Banca:")
for dia, valor_banca in enumerate(historico_banca, 1):
    print(f"Dia {dia}: R${valor_banca:.2f}")

print(f"\nBanca final após {dias} dias: R${historico_banca[-1]:.2f}")
```
