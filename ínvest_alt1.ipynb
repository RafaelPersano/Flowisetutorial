```python
import matplotlib.pyplot as plt
import numpy as np

def solicitar_dados_usuario():
    print("=== Entrada de Dados ===")
    investimento = float(input("Qual é o valor total disponível para investir? R$ "))
    tempo = int(input("Em quantos meses você deseja obter retorno (ex: 3, 6, 12)? "))
    risco = input("Qual é o seu perfil de risco (Baixo, Médio, Alto)? ").lower()
    experiencia = input("Qual é o seu nível de experiência em investimentos (Iniciante, Intermediário, Avançado)? ").lower()
    objetivo = input("Qual é o seu principal objetivo (Aumentar Renda, Preservar Capital, Especular)? ").lower()
    return investimento, tempo, risco, experiencia, objetivo

def calcular_probabilidades(investimento, tempo, risco, experiencia, objetivo):
    base_prob = 0.5

    if risco == "alto":
        base_prob += 0.2
    elif risco == "médio":
        base_prob += 0.1

    if experiencia == "avançado":
        base_prob += 0.1
    elif experiencia == "iniciante":
        base_prob -= 0.1

    if objetivo == "especular":
        base_prob += 0.1
    elif objetivo == "preservar capital":
        base_prob -= 0.1

    probabilidade = max(0.2, min(base_prob, 0.9))
    return round(probabilidade, 2)

def gerar_cronograma(estrategia, probabilidade):
    passos = []

    if probabilidade > 0.7:
        passos.append("Curto Prazo: Execute ações iniciais para validação e ajustes.")
        passos.append("Médio Prazo: Escale a estratégia com diversificação moderada.")
        passos.append("Longo Prazo: Automatize processos e avalie trimestralmente.")
        passos.append("Longo Prazo: Invista em ferramentas de análise para maximizar ganhos.")
    elif probabilidade > 0.4:
        passos.append("Curto Prazo: Invista em aprendizado e simulações.")
        passos.append("Curto Prazo: Comece com um pequeno percentual do seu capital.")
        passos.append("Médio Prazo: Diversifique riscos e busque feedback de especialistas.")
        passos.append("Longo Prazo: Reavalie constantemente e ajuste suas metas.")
    else:
        passos.append("Curto Prazo: Evite investimentos de alto risco e foque em aprendizado.")
        passos.append("Curto Prazo: Construa um fundo de emergência antes de avançar.")
        passos.append("Médio Prazo: Construa um portfólio conservador.")
        passos.append("Longo Prazo: Mantenha disciplina e estude novas oportunidades.")

    return passos

def calcular_retorno_mensal(investimento, probabilidade, tempo):
    retornos = []
    taxa_retorno = probabilidade * 0.05  # Taxa de retorno baseada na probabilidade

    for mes in range(1, tempo + 1):
        investimento *= (1 + taxa_retorno)
        retornos.append(investimento)

    return retornos

def gerar_graficos(probabilidade, retornos, tempo):
    categorias = ['Sucesso', 'Risco']
    valores = [probabilidade, 1 - probabilidade]

    plt.figure(figsize=(12, 6))

    # Gráfico de pizza
    plt.subplot(1, 2, 1)
    plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=140, colors=['#4CAF50', '#F44336'])
    plt.title("Probabilidade de Sucesso")

    # Gráfico de linha
    plt.subplot(1, 2, 2)
    meses = np.arange(1, tempo + 1)
    plt.plot(meses, retornos, marker='o', color='#4CAF50', label="Retorno Mensal")
    plt.xlabel("Meses")
    plt.ylabel("Valor do Investimento (R$)")
    plt.title("Evolução do Retorno do Investimento")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

def main():
    investimento, tempo, risco, experiencia, objetivo = solicitar_dados_usuario()
    probabilidade = calcular_probabilidades(investimento, tempo, risco, experiencia, objetivo)

    print(f"\nProbabilidade calculada de sucesso: {probabilidade * 100:.1f}%")

    estrategias = {
        "Apostas Esportivas": "Explorar mercados estáveis como futebol e tênis.",
        "Criação de Tokens": "Desenvolver um token simples na testnet com usabilidade clara.",
        "Investimentos Alternativos": "Diversificar em setores emergentes como startups e economia verde."
    }

    for estrategia, descricao in estrategias.items():
        print(f"\nEstratégia: {estrategia}")
        print(f"Descrição: {descricao}")

        passos = gerar_cronograma(estrategia, probabilidade)
        for i, passo in enumerate(passos, 1):
            print(f"Passo {i}: {passo}")

    retornos = calcular_retorno_mensal(investimento, probabilidade, tempo)
    gerar_graficos(probabilidade, retornos, tempo)

    print("\nRecomenda-se seguir o cronograma sugerido para maximizar as chances de sucesso.")

if __name__ == "__main__":
    main()
```
