```python
import matplotlib.pyplot as plt
import numpy as np

def solicitar_dados_usuario():
    """Solicita dados do usuário sobre investimento, tempo, risco, experiência e objetivo."""
    print("=== Entrada de Dados ===")
    while True:
        try:
            investimento = float(input("Qual é o valor total disponível para investir? R$ "))
            if investimento <= 0:
                raise ValueError("O valor do investimento deve ser positivo.")
            break
        except ValueError as e:
            print(f"Valor de investimento inválido: {e}. Por favor, insira um número positivo.")

    while True:
        try:
            tempo = int(input("Em quantos meses você deseja obter retorno (ex: 3, 6, 12)? "))
            if tempo <= 0:
                raise ValueError("O tempo deve ser positivo.")
            break
        except ValueError as e:
            print(f"Valor de tempo inválido: {e}. Por favor, insira um número positivo.")

    while True:
        risco = input("Qual é o seu perfil de risco (Baixo, Médio, Alto)? ").lower()
        if risco in ["baixo", "médio", "alto"]:
            break
        else:
            print("Perfil de risco inválido. Por favor, escolha entre Baixo, Médio ou Alto.")

    while True:
        experiencia = input("Qual é o seu nível de experiência em investimentos (Iniciante, Intermediário, Avançado)? ").lower()
        if experiencia in ["iniciante", "intermediário", "avançado"]:
            break
        else:
            print("Nível de experiência inválido. Por favor, escolha entre Iniciante, Intermediário ou Avançado.")

    while True:
        objetivo = input("Qual é o seu principal objetivo (Aumentar Renda, Preservar Capital, Especular)? ").lower()
        if objetivo in ["aumentar renda", "preservar capital", "especular"]:
            break
        else:
            print("Objetivo inválido. Por favor, escolha entre Aumentar Renda, Preservar Capital ou Especular.")

    return investimento, tempo, risco, experiencia, objetivo

def calcular_probabilidades(risco, experiencia, objetivo):
    """Calcula a probabilidade de sucesso com base no perfil do usuário."""
    base_prob = 0.5

    if risco == "alto":
        base_prob += 0.2
    elif risco == "médio":
        base_prob += 0.1

    if experiencia == "avançado":
        base_prob += 0.1
    elif experiencia == "iniciante":
        base_prob -= 0.1

    if objetivo == "especular":
        base_prob += 0.1
    elif objetivo == "preservar capital":
        base_prob -= 0.1

    probabilidade = max(0.2, min(base_prob, 0.9))
    return round(probabilidade, 2)

def gerar_cronograma(estrategia, probabilidade, risco, objetivo):
    """Gera um cronograma de investimento com base na estratégia, probabilidade, risco e objetivo."""
    passos = []

    if probabilidade > 0.7:
        passos.extend([
            "Curto Prazo: Execute ações iniciais para validação e ajustes.",
            "Médio Prazo: Escale a estratégia com diversificação moderada.",
            "Longo Prazo: Automatize processos e avalie trimestralmente.",
            "Longo Prazo: Invista em ferramentas de análise para maximizar ganhos."
        ])
    elif probabilidade > 0.4:
        passos.extend([
            "Curto Prazo: Invista em aprendizado e simulações.",
            "Curto Prazo: Comece com um pequeno percentual do seu capital.",
            "Médio Prazo: Diversifique riscos e busque feedback de especialistas.",
            "Longo Prazo: Reavalie constantemente e ajuste suas metas."
        ])
    else:
        passos.extend([
            "Curto Prazo: Evite investimentos de alto risco e foque em aprendizado.",
            "Curto Prazo: Construa um fundo de emergência antes de avançar.",
            "Médio Prazo: Construa um portfólio conservador.",
            "Longo Prazo: Mantenha disciplina e estude novas oportunidades."
        ])

    # Ajustes no cronograma com base no risco e objetivo
    if risco == "baixo" and objetivo == "preservar capital":
        passos.append("Priorize investimentos de baixo risco, como títulos do governo.")
    elif risco == "alto" and objetivo == "especular":
        passos.append("Considere alocar uma pequena parte do capital em investimentos de alto risco.")

    return passos

def calcular_retorno_mensal(investimento, probabilidade, tempo):
    """Calcula o retorno mensal projetado."""
    retornos = []
    taxa_retorno = probabilidade * 0.05  # Taxa de retorno baseada na probabilidade

    for mes in range(1, tempo + 1):
        investimento *= (1 + taxa_retorno)
        retornos.append(investimento)

    return retornos

def gerar_graficos(probabilidade, retornos, tempo):
    """Gera gráficos de probabilidade de sucesso e evolução do retorno."""
    categorias = ['Sucesso', 'Risco']
    valores = [probabilidade, 1 - probabilidade]

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=140, colors=['#4CAF50', '#F44336'])
    plt.title("Probabilidade de Sucesso")

    plt.subplot(1, 2, 2)
    meses = np.arange(1, tempo + 1)
    plt.plot(meses, retornos, marker='o', color='#4CAF50', label="Retorno Mensal")
    plt.xlabel("Meses")
    plt.ylabel("Valor do Investimento (R$)")
    plt.title("Evolução do Retorno do Investimento")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()

def main():
    """Função principal que executa o programa."""
    investimento, tempo, risco, experiencia, objetivo = solicitar_dados_usuario()
    probabilidade = calcular_probabilidades(risco, experiencia, objetivo)

    print(f"\nProbabilidade calculada de sucesso: {probabilidade * 100:.1f}%")

    estrategias = {
        "Apostas Esportivas": "Explorar mercados estáveis como futebol e tênis.",
        "Criação de Tokens": "Desenvolver um token simples na testnet com usabilidade clara.",
        "Investimentos Alternativos": "Diversificar em setores emergentes como startups e economia verde.",
        "Ações de Valor": "Investir em empresas subvalorizadas com potencial de crescimento.",
        "Fundos Imobiliários": "Diversificar em imóveis através de fundos imobiliários."
    }

    for estrategia, descricao in estrategias.items():
        print(f"\nEstratégia: {estrategia}")
        print(f"Descrição: {descricao}")

        passos = gerar_cronograma(estrategia, probabilidade, risco, objetivo)
        for i, passo in enumerate(passos, 1):
            print(f"Passo {i}: {passo}")

    retornos = calcular_retorno_mensal(investimento, probabilidade, tempo)
    gerar_graficos(probabilidade, retornos, tempo)

    print("\nRecomenda-se seguir o cronograma sugerido para maximizar as chances de sucesso.")

if __name__ == "__main__":
    main()
```
