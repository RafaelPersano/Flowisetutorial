```python
import random

# Configurações iniciais
initial_bankroll = 1000.00  # Banca inicial
daily_goal_percentage = 5  # Meta diária (%)
days_to_simulate = 30  # Número de dias para simular
players = ["LeBron James", "Stephen Curry", "Kevin Durant", "Giannis Antetokounmpo", "Luka Dončić"]  # Lista de jogadores fictícios

def generate_random_player_stats():
    """
    Gera estatísticas aleatórias para os jogadores.
    """
    player_stats = []
    for player in players:
        stats = {
            "Name": player,
            "Points": random.uniform(20, 35),  # Pontuação média entre 20 e 35 pontos
            "Rebounds": random.uniform(5, 15),  # Rebotes entre 5 e 15
            "Assists": random.uniform(3, 10)  # Assistências entre 3 e 10
        }
        player_stats.append(stats)
    return player_stats

def suggest_bets(player_stats):
    """
    Sugere uma aposta com base nos dados aleatórios de estatísticas dos jogadores.
    """
    player = random.choice(player_stats)  # Escolhe um jogador aleatório
    player_name = player["Name"]
    avg_points = player["Points"]
    suggestion = f"Aposte que {player_name} fará mais de {avg_points * 0.9:.1f} pontos."
    return suggestion

def simulate_betting(bankroll, daily_goal, days):
    """
    Simula o processo de apostas para alcançar uma meta diária durante o período de simulação.
    """
    print(f"Simulação de {days} dias")
    print(f"Banca inicial: R$ {bankroll:.2f}")
    print(f"Meta diária: R$ {daily_goal:.2f}\n")

    for day in range(1, days + 1):
        print(f"--- Dia {day} ---")
        print("Gerando estatísticas aleatórias para os jogadores...")

        player_stats = generate_random_player_stats()  # Gera estatísticas fictícias
        bet_suggestion = suggest_bets(player_stats)
        print(f"Sugestão de aposta: {bet_suggestion}")

        # Simula o resultado da aposta
        bet_result = random.choice(["win", "loss"])  # Ganha ou perde (50% de chance)
        if bet_result == "win":
            bankroll += daily_goal
            print(f"Parabéns! Você atingiu a meta diária. Nova banca: R$ {bankroll:.2f}\n")
        else:
            bankroll -= daily_goal / 2  # Perda de 50% do valor diário
            print(f"Infelizmente você perdeu. Nova banca: R$ {bankroll:.2f}\n")

        # Verifica se a banca foi zerada
        if bankroll <= 0:
            print("Sua banca foi zerada. Simulação encerrada.")
            break

    print(f"Simulação encerrada. Banca final: R$ {bankroll:.2f}")

# Calcula a meta diária
daily_goal = initial_bankroll * (daily_goal_percentage / 100)

# Executa a simulação
simulate_betting(initial_bankroll, daily_goal, days_to_simulate)
```
